```
#clean the address state column
from pyspark.sql.functions import col , regexp_extract,upper, isnan, isnull, count, log, when
lc_df=lc_df.withColumn("addr_state",regexp_extract(upper(col("addr_state")),"[a-zA-Z]{2}",0))
```

```
lc_df.show(10)
```

```
lc_df.createOrReplaceTempView("lc_loan_data")
```
